```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from sklearn import linear_model
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor
from sklearn.preprocessing import PolynomialFeatures
from  sklearn.linear_model import LinearRegression
```

```python
from sklearn.datasets import load_boston
data  = load_boston()
print(data.DESCR)
```

```python
data.data.shape
```

```python
boston =load_boston()
print(boston.data.shape, boston.target.shape)
print(boston.feature_names)
```

```python
data = pd.DataFrame(boston.data, columns = boston.feature_names)
data = pd.concat([data, pd.Series(boston.target, name = 'MEDV')], axis = 1)
data.head()
```

```python
data.columns
```

```python
x=data[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
y= boston.target
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()
predictions=model.predict(x)
model.summary()
```

```python
poly= PolynomialFeatures(degree=2)
x=poly.fit_transform(x)
model2 = linear_model.LinearRegression()
model2.fit(x,y)
r_squared=model2.score(x,y)
adjusted_r_squared = 1-(1-r_squared)*(len(y)-1)/(len(y)-x.shape[1]-1)
print('Polynomial Model of degree 2 -R square is %.2f R adj square %.2f'%(r_squared,adjusted_r_squared))
```

```python
poly= PolynomialFeatures(degree=3)
x=poly.fit_transform(x)
model2 = linear_model.LinearRegression()
model2.fit(x,y)
r_squared=model2.score(x,y)
adjusted_r_squared = 1-(1-r_squared)*(len(y)-1)/(len(y)-x.shape[1]-1)
print('Polynomial Model of degree 3 -R square is %.2f R adj square %.2f'%(r_squared,adjusted_r_squared))
```
